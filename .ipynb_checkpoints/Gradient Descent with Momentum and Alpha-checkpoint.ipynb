```python
%matplotlib inline
import matplotlib.pylab as plt


import time
from IPython import display
import numpy as np

y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
x = np.array([10., 8., 13., 9., 11., 14., 6., 4., 12., 7., 5.])
N = len(x)

# Design matrix
A = np.vstack((np.ones(N), x, x**2, x**3)).T # 3.dereece # ax^3 + bx^2 + cx + d

w_best, E, rank, s = np.linalg.lstsq(A, y) # Least Squares -> sondaki least squaresten dönen sonucu diğer variablelara dağıtıyor
err = y - A.dot(w_best)
E_min = np.sum(err**2)/N

w = np.array([2., 1., 0.1, 0.001]) #a,b,c,d
p = np.zeros(4)

EPOCHS = 100
W = np.zeros((4,EPOCHS))

alpha = 0.00001
beta = 0.2

for epoch in range(EPOCHS):
    err = y-A.dot(w)
    W[:,epoch] = w

    E = np.sum(err**2) / N
    
    dE = -2. * A.T.dot(err) / N # türev
    p = dE + beta * p

    if epoch % 1 == 0: 
        print(epoch,':', E)

    w = w - alpha * p

```
